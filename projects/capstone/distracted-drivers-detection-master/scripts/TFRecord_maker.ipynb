## Writing `.tfrecord` file

TF-slim provides several useful functions that will help us to creat TFRecord file without dig into the low-level TensorFlow. KWOT SIN also compiles necessary functions into `dataset_utils.py` file. 

First import required modules.

````python
import random
import tensorflow as tf
````

````python
from datasets.dataset_utils import _dataset_exists, _get_filenames_and_classes, _get_test_filenames, write_label_file, _convert_dataset 
````

And required arguments:

````python
train_dir = '../dataset_original/train'

test_dir = '../dataset_original/test'

tf_dir = '../dataset_tf'

if not tf.gfile.Exists(tf_dir):
    tf.gfile.MakeDirs(tf_dir)

validation_size = 0.2

num_shards_train = 20

num_shards_val = 10

num_shards_test = 40

random_seed = 42

tfrecord_filename = 'drivers'
````

Check if the TFRecord files exist.

````python
if _dataset_exists(dataset_dir=tf_dir, 
                   _NUM_SHARDS = [num_shards_train, num_shards_val, num_shards_test], 
                   output_filename = tfrecord_filename):
    print('Dataset files already exist.')
````

Write the `.tfrecord` files.

get image filenames

````python
train_imgs, class_names = _get_filenames_and_classes(train_dir)
test_imgs = _get_test_filenames(test_dir)
````

````python
class_names_to_ids = dict(zip(class_names, range(len(class_names))))
````

````python
#Find the number of validation examples we need
num_validation = int(validation_size * len(train_imgs))

# Divide the training datasets into train and test:
random.seed(random_seed)
random.shuffle(train_imgs)
training_filenames = train_imgs[num_validation:]
validation_filenames = train_imgs[:num_validation]
````

````python
# First, convert the training and validation sets.
_convert_dataset('train', training_filenames, class_names_to_ids,
                 dataset_dir = tf_dir,
                 tfrecord_filename = tfrecord_filename,
                 _NUM_SHARDS = num_shards_train)
_convert_dataset('validation', validation_filenames, class_names_to_ids,
                 dataset_dir = tf_dir,
                 tfrecord_filename = tfrecord_filename,
                 _NUM_SHARDS = num_shards_val)
_convert_dataset('test', test_imgs, class_names_to_ids,
                 dataset_dir = tf_dir,
                 tfrecord_filename = tfrecord_filename,
                 _NUM_SHARDS = num_shards_test)
````

````python
# Finally, write the labels file:
labels_to_class_names = dict(zip(range(len(class_names)), class_names))
write_label_file(labels_to_class_names, tf_dir)
````

## read `.tfrecord`

````python
from skimage.io import imread, imshow
from dataset_preparation import get_split, load_batch
from preprocessing.preprocessing_factory import get_preprocessing
from nets import nets_factory
%matplotlib inline  
````

````python
file_pattern = 'drivers_{}_*.tfrecord'
````

````python
MODEL = 'resnet_v2_50'
model = nets_factory.get_network_fn(MODEL, 10, is_training=False)
image_size = model.default_image_size
````

````python
def load_tf(split_name): 
    with tf.Graph().as_default() as graph:
        tf.logging.set_verbosity(tf.logging.INFO)

        dataset = get_split(split_name=split_name, 
                            dataset_dir=tf_dir, 
                            file_pattern=file_pattern, 
                            file_pattern_for_counting=tfrecord_filename, 
                            items_to_descriptions=None)
        image, image_raw, label, image_name = load_batch(
            dataset=dataset,
            batch_size=1, 
            MODEL=MODEL,
            height=image_size,
            width=image_size,
            is_training=False)

        sv = tf.train.Supervisor(logdir='../log/tf_maker')

        with sv.managed_session() as sess:
            for i in range(dataset.num_samples):
                images, images_raw, labels, image_names = sess.run([image, image_raw, label, image_name])
                img, img_raw, lbl, img_n = images[0], images_raw[0], labels[0], image_names[0]
                print(images.shape)
                print(images_raw.shape)
                print(lbl)
                print(img_n)
                imshow(img_raw)
                break        
````

````python
load_tf('train')
````

````python
load_tf('validation')
````

````python
load_tf('test')
````
